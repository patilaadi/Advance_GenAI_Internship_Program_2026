# 1: Mobile Recharge Validation System


```python
# function to validate mobile recharge
def mobile_recharge(amount):
    
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge request rejected..! Recharge plan must be greater than 50Rs"
    
    if amount not in valid_plans:
        return "Enter a valid plan (Ex. 199, 299, 399, 599)"
    
    return f"Recharge Request Accepted for Rs. {amount}"


# user input loop
while True:
    print("-------Mobile Recharge Request---------")
    amount = int(input("Enter Recharge amount: "))
    
    result = mobile_recharge(amount)
    print(result)
    
    if "Accepted" in result:
        break
```

#  2: Inventory Reorder Alert System

```python
# function for product reorder
def check_stock(product_dict):
    
    for product, quantity in product_dict.items():
        
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# Dictionary of products
products = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Wheat": 18
}

# Call function
check_stock(products)
```

# 3: Student Result Processing 

```python
#function for check student result
def check_result(marks):
    
    average = sum(marks) / len(marks)
    
    print(f"\nAverage Marks: {average}")
    
    if average >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")


# Taking input from user
marks = []

n = int(input("Enter number of subjects: "))

for i in range(n):
    mark = float(input(f"Enter marks for subject {i+1}: "))
    marks.append(mark)

# Call function
check_result(marks)
```

# 4: Cab Fare Estimator with Retry Option

```python
# Function to calculate fare
def calculate_fare(distance, peak):
    
    base_fare = 50
    per_km_rate = 12
    
    total_fare = base_fare + (distance * per_km_rate)
    
    if peak.lower() == "yes":
        total_fare += total_fare * 0.25   # 25% extra
    
    return total_fare


# While loop for retry
while True:
    
    print("\n------ Cab Fare Calculator ------")
    
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")
    
    fare = calculate_fare(distance, peak)
    
    print(f"Total Fare: ₹{fare}")
    
    retry = input("Do you want to calculate again? (yes/no): ")
    
    if retry.lower() != "yes":
        print("Thank you for using Cab Fare Calculator")
        break
```

# 5: Employee Attendance Eligibility Checker


```python
#sample attendance
attendance_list = ["P", "A", "P", "P", "P", "A","P","P", "P","P"]

#function for checking employee eligibility
def employee_eligibility(attendance_list):
    present_count = 0
    
    # Count present days
    for status in attendance_list:
        if status == "P":
            present_count += 1

     # Calculate percentage
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

     # Check eligibility
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% - Eligible"
    else:
        return f"Attendance: {percentage:.2f}% - Not Eligible"

#function use
result = employee_eligibility(attendance_list)
print(result)
```

# 6: Password Strength Checker


```python
# function for check password length
def check_password(password):
    
    # Minimum length
    if len(password) < 8:
        return "Weak Password: Must be at least 8 characters long"
    
    has_digit = False
    has_special = False
    
    # Loop through each character
    for char in password:
        
        if char.isdigit():
            has_digit = True
        
        if char in "@#$":
            has_special = True
    
    # Check conditions
    if not has_digit:
        return "Weak Password: Must contain at least one digit"
    
    if not has_special:
        return "Weak Password: Must contain at least one special character (@#$)"
    
    return "Strong Password "


# function usage
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
