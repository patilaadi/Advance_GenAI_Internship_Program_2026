# 1: Unique Words in a Sentence

```python


# Taking input from user
sentence = input("Enter sentence: ")

# Step 1: Split sentence into words
words = sentence.split()

# Step 2: Convert list to set (removes duplicates)
unique_words = set(words)

# Step 3: Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)

     
```

# 2. Highest Salary from Employee Data

```python
# Employee dictionary
employees = {
    "Amar": 90000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

# Print result
print("Highest Salary:", highest_employee, "-", employees[highest_employee])

```

# 3.Find Maximum and Minimum Values

```python
# Given list
numbers = [55,69,58,78,96,26]

# Step 1: Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Step 2: Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Step 3: Print results
print("List:", numbers)
print("Maximum Value from list:", maximum)
print("Minimum Value from list:", minimum)
```

# 4.Count Products Above a Price Threshold

```python

# Given list of product prices
prices = [550, 1300, 899, 1400, 1400]

# Counter variable
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
     
```

# 5. Calculate Attendance Percentage

```python
# Given attendance list
attendance = ["P", "P", "A", "A", "A"]

# Step 1: Count total days
total_days = len(attendance)

# Step 2: Count number of Present days
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

# Step 3: Calculate percentage
attendance_percentage = (present_count / total_days) * 100

# Step 4: Print result
print("Attendance Percentage:", attendance_percentage)

     
```

# 7. Count Character Frequency

```python


# Given string
text = "Dictionary"

# Empty dictionary to store frequency
char_count = {}

# Loop through each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

# Print result
print(char_count)

     
```

# 9. Check if a Key Exists in a Dictionary

```python
# Given dictionary
employees = {
    "Ankit": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_to_check = "Ravi"

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

# 10.  Calculate Average Marks

```python
# Given list of marks
marks = [65, 80, 85, 75]

# Step 1: Calculate total marks
total = sum(marks)

# Step 2: Calculate number of subjects
count = len(marks)

# Step 3: Calculate average
average = total / count

# Print result
print("Average Marks:", average)
```
