# Online shopping cart validation

```python
#no of products
n= int (input("enter no. of products"))
#empty cart 
cart_prices=[]
#loop
for i in range(n):
    price =float(input(f"Enter price of product{i+1}:"))
    cart_prices.append(price)
#intialize total
total=0
for price in cart_prices:
    if price<=0:
        print("invalid product price")
        break
    total= total+ price
else:
    print("Cart Total:",total)
    if total>2000:
        discount = total *0.10
        total=total-discount
        print("Discount applied:",discount)
    if total>500:
        print("Final payble amount:",total)
    else:
        print("Add more product to get discount")
```
