## 1.Employee Performance Bonus Eligibility

```python
#Given data
employees={"Ravi:":92,
            "Ankita":88,
            "Kiran":92,
            "Suresh":85
          }
#highest performance score
highest_score = max(employees.values())

#handling same highest score
top_employees = [name for name, score in employees.items() if score == highest_score]

print(f"Top Performers Eligible for Bonus: {top_employees} (Score:{highest_score})")
```

## 2. Search Query Keyword Analysis

```python
import string
#Given text
text="Buy mobile phone buy phone online"

#convert input into lowercase
lower_text=text.lower()

#ignoring common punctuation
text = lower_text.translate(str.maketrans('', '', string.punctuation))

#count the frequency of each keyword
words = text.split()
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1


#Display only keywords searched more than once
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

## 3. Sensor Data Valiation

```python
#Given Sensor readings
sensor_readings = [3,4,7,8,10,12,5]

#Identify readings that are even numbers 
valid_readings = [(index, value) 
                  for index, value in enumerate(sensor_readings) 
                  if value % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## 4. Email Domain Usage Analysis

```python
#Given inputs
emails=[
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

#Count how many users belong to each email domain
domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
   

#Calculate the percentage usage of each domain
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

## 5. Sales Spike Detection

```python
import math
#Given Data
sales=[1200, 1500, 900, 2200, 1400, 3000]

#Calculate average daily sales
avg_sales=sum(sales) / len(sales)

#Detect days where sales are more than 30% above average
for index,value in enumerate(sales):
    if value > avg_sales:
        print(f"Day {index + 1}: {value}")
    
```

## 6. Duplicate User ID Detection

```python
#given Data
user_ids= ["user1","user2","user1","user3","user1","user3"]

#Identify duplicate user IDs
count_dict={}

for users in user_ids:
    count_dict[users]= count_dict.get(users,0)+1

#Display how many times each duplicate appears.
for user, count in count_dict.items():
    if count > 1:
        print(f"{user} → {count} times")
```
