## 1. Mobile Recharge Validation System

```python
# function to validate mobile recharge
def mobile_recharge(amount):
    
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge request rejected..! Recharge plan must be greater than 50Rs"
    
    if amount not in valid_plans:
        return "Enter a valid plan (Ex. 199, 299, 399, 599)"
    
    return f"Recharge Request Accepted for Rs. {amount}"


# user input loop
while True:
    print("-------Mobile Recharge Request---------")
    amount = int(input("Enter Recharge amount: "))
    
    result = mobile_recharge(amount)
    print(result)
    
    if "Accepted" in result:
        break
```
