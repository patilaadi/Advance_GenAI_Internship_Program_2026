```python
def manage_parking(capacity, logs):
    
    current = 0
    
    # Process each log entry
    for event in logs:
        if event == "IN":
            current = current + 1
        elif event == "OUT":
            current = current - 1
            # Ensure we don't go below zero
            if current < 0:
                current = 0
    
    # Checking if parking is full
    if current >= capacity:
        status = "FULL"
    else:
        status = "Available"
    
    # Printing results
    print(f"Currently Parked Vehicles: {current}")
    print(f"Parking Status: {status}")

# Testing with sample input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, vehicle_logs)
```

```python
def estimate_delivery_time(distance, traffic_level, weather):
    # For 8 km, base time = 8 * 3.75 = 30 minutes
    base_time = distance * 3.75
    
    # Traffic adjustments
    if traffic_level == "Low":
        traffic_factor = 1.0  # 0% increase
    elif traffic_level == "Medium":
        traffic_factor = 1.3  # 30% increase
    elif traffic_level == "High":
        traffic_factor = 1.4  # 40% increase
    
    # Weather adjustments  
    if weather == "Clear":
        weather_factor = 1.0  # 0% increase
    elif weather == "Rainy":
        weather_factor = 1.3  # 30% increase
    elif weather == "Stormy":
        weather_factor = 1.6  # 60% increase
    elif weather == "Snowy":
        weather_factor = 2.0  # 100% increase
    
    # Calculate final time
    # First apply traffic, then weather (or combine them)
    time_after_traffic = base_time * traffic_factor
    final_time = time_after_traffic * weather_factor
    
    # Round to nearest minute
    final_time = round(final_time)
    
    print(f"Estimated Delivery Time: {final_time} minutes")
    return final_time

# Test with sample input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
def seat_analyzer(total_seats, booked_seats):
    # Calculate occupancy
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    occupancy = round(occupancy)
    
    # Determine status
    if occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 50:
        status = "Partial"
    else:
        status = "Low"
    
    # Print in expected format
    print(f"Occupancy: {occupancy}%")
    print(f"Show Status: {status}")
    
    # Additional logic for opening shows
    if occupancy >= 90:
        print("Suggestion: Open additional show")
    elif occupancy >= 75:
        print("Suggestion: Consider opening another show")

# Testing
seat_analyzer(200, [1] * 150)
```

```python
def classify_server_load(cpu_readings):
    # Calculate average CPU load
    total = 0
    for reading in cpu_readings:
        total = total + reading
    
    average_load = total / len(cpu_readings)
    average_load = round(average_load)
    
    # Determine server status based on average load
    if average_load < 50:
        status = "Normal"
    elif average_load >= 50 and average_load < 80:
        status = "Warning"
    else:  # average_load >= 80
        status = "Critical"
    
    # Display results
    print(f"Average CPU Load: {average_load}%")
    print(f"Server Status: {status}")
    
    return average_load, status

# Test with sample input
cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```

```python
def monitor_classroom_resources(usage_data):
    # Define usage limits (maximum recommended hours)
    limits = {
        "Projector": 5,
        "AC": 8,
        "Lights": 10
    }
    
    overused_resources = []
    energy_alert = "No"
    
    # Checking each resource for overuse
    for resource, hours in usage_data.items():
        if resource in limits:
            if hours > limits[resource]:
                overused_resources.append(resource)
    
    # Set energy alert if any resource is overused
    if len(overused_resources) > 0:
        energy_alert = "Yes"
    
    # Display results
    if len(overused_resources) > 0:
        print(f"Overused Resources: {', '.join(overused_resources)}")
    else:
        print("Overused Resources: None")
    
    print(f"Energy Alert: {energy_alert}")
    
    return overused_resources, energy_alert

# Test with sample input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_classroom_resources(usage_data)
```

```python
def manage_event_registrations(capacity, registration_count):
    
    # Check if registrations exceed capacity
    if registration_count <= capacity:
        confirmed = registration_count
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registration_count - capacity
        status = "Closed"
    
    # Display results
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")
    
    return confirmed, waitlisted, status

# Test with sample input
capacity = 100
registrations = 105

manage_event_registrations(capacity, registrations)
```
